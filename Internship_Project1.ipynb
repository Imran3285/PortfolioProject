```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

#here is the sample data

data = {
    'Day' : ['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday'],
    'Temperature': [20, 25, 22, 30, 28, 32, 18],
    'Sunny' : ['No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No'],
    'Sales' : [30, 50, 35, 70, 60, 80, 20]
}
df = pd.DataFrame(data)

print("Dataset Overview:\n")
print(df.info())
print("\nPreview:\n")
print(df.head())
print("\nDescriptive Stats:\n")
print(df.describe())

#calculate Descriptive Statistics

mean_temp = df['Temperature'].mean()
median_temp = df['Temperature'].median()
std_temp = df['Sales'].std()
mean_sales = df['Sales'].mean()
median_sales = df['Sales'].median()
std_sales = df['Sales'].std()
print(f"Temperature - Mean: {mean_temp:.2f}, Median: {median_temp:.2f}, Std Dev: {std_temp:.2f}")
print(f"Sales- Mean: {mean_sales:.2f}, Median: {median_sales:.2f}, Std Dev: {std_sales:.2f}")

#visualisng data histogram

plt.figure(figsize=(8,5))
sns.scatterplot(x='Temperature', y='Sales', hue='Sunny', data=df)
plt.title("Sales vs. Temperature")
plt.xlabel("Temperature(Celsius)")
plt.ylabel("Sales(cups)")
plt.show()

"""
The scatter plot shows how sales rise positively with temperature. As the temperature rise more lemoade is sold.
Hence, the demand is weather sensitive, and better weather forecast can maximize profit.
"""
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Sunny', y='Sales', data=df)
plt.title("Average Sales by Weather Conditions")
plt.xlabel('Sunny (Yes/No)')
plt.ylabel('Average Sales(cups)')
plt.show()
```

```python
Correlation = df['Temperature'].corr(df['Sales'])
print(f"Correlation between Temperature and Sales: {Correlation:.2f}")
#In here the correlation efficient is 1 that shows very strong positive correlation.
```

```python
print("\n Business Insights & Recommendations:\n")
print("- Sales increase strongly with temperature (R close to 1).")
print("- Sunny days boost sales compared to non-sunny days.")
print("- Store should stock more lemonade and staff more workers on hot & sunny days.")
print("- Weather forecasts can be used for inventory planning to avoid shortages or waste.")
```
